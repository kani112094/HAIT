```python
#ライブラリの読み込み
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline #欠損値NaN
from numpy import nan
#サンプルデータの作成
df = pd.DataFrame({'DATE':  [20170801, 20170801, 20170802, 20170802, 20170803, 20170803, 20170803, 20170805, 20170805, 20170805],
                   'NECK':  [41, nan, 38, 46, nan, 37, nan, 38, nan, 42],
                   'BODY':  [84, 92, nan, 90, nan, 64, 78, 74, 82, 86],
                   'SIZE':  ['L', 'XL', 'L', 'XL', 'M', 'S', 'M', 'L', 'L', 'XL'],
                   'COLOR': ['BL', 'RD', 'Y', 'GR', 'GR', 'RD', 'BL', 'Y', 'BL', 'GR'],
                   'class': ['A', 'C', 'B', 'B', 'C', 'A', 'A', 'A', 'C', 'C']},
                 columns=['DATE', 'NECK', 'BODY', 'SIZE', 'COLOR', 'class'])
df
```

```python
#欠損値をカウント
df.isnull().sum()
```

```python
#欠損値を含むサンプルを削除(非破壊)←リストワイズ法
df.dropna()
```

```python
#特定の特徴量に欠損のあるサンプルを削除(非破壊)
df.dropna(subset=["BODY"])
```

```python
#欠損していない値が5つ以上あるサンプルを残す(非破壊)
df.dropna(thresh=5)
```

```python
#平均値補完(非破壊)
df.fillna(df.mean())
```

```python
#任意の値で補完(非破壊)
df.fillna(0)
```

```python
#回帰補完(非破壊)
df.interpolate(method="linear")
```

```python
#ライブラリの読み込み
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
from numpy import nan
#サンプルデータの作成
df = pd.DataFrame({'DATE':  [20170801, 20170801, 20170802, 20170802, 20170803, 20170803, 20170803, 20170805, 20170805, 20170805],
                   'NECK':  [41, nan, 38, 46, nan, 37, nan, 38, nan, 42],
                   'BODY':  [84, 92, nan, 90, nan, 64, 78, 74, 82, 86],
                   'SIZE':  ['L', 'XL', 'L', 'XL', 'M', 'S', 'M', 'L', 'L', 'XL'],
                   'COLOR': ['BL', 'RD', 'Y', 'GR', 'GR', 'RD', 'BL', 'Y', 'BL', 'GR'],
                   'class': ['A', 'C', 'B', 'B', 'C', 'A', 'A', 'A', 'C', 'C']},
                 columns=['DATE', 'NECK', 'BODY', 'SIZE', 'COLOR', 'class'])
df
```

```python
#順序特徴量(シャツのサイズ)と整数を対応させる辞書を作成
size_mapping={"S":1,"M":2,"L":3,"XL":4}
#マッピングの実行
df['SIZE'] = df['SIZE'].map(size_mapping)
df
```

```python
#名義特徴量(シャツの色)のダミー変数を作成
pd.get_dummies(df["COLOR"])
```

```python
#クラスラベルと整数を対応させる辞書を作成
class_mapping={"A":0,"B":1,"C":2}
#マッピングの実行
df["class"]=df["class"].map(class_mapping)
df
```
